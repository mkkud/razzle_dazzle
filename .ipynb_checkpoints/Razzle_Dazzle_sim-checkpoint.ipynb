# Simplified Razzle Dazzle simulator 

```python
import random, numpy
```

```python
def razzle_throw():
    score = 0
    dice = 0
    while dice < 8:
        dice += 1 
        score += random.randint(1, 6)
    else:
        return score
```

```python
def razzle_game(num_tries = 100, my_money = 100, prize_value = 100, prize_points = 100):
    '''
    Simulates a simplified RAZZLE DAZZLE game. 
    
    Your initial money is specified as my_money.
    You ask for num_tries of tries, each costs a dollar (at start). 
    You aim to get the prize, which value is prize_value.
    
    The rules are simple; if your number on the razzle board is red, you get the specified number points. 
    If it's black, you get nothing.
    If it's green, the prize value is increased by $100 (Initial prize is $100), but its cost stays the same. 
    You get the prize if you get up to prize_points points.
    
    But beware! If you roll 29, your payment doubles!
    
    This game is so rigged against the player that it's illegal in the US. Have fun!
    '''
    my_money = my_money
    green = (18, 38, 19, 37, 36, 21, 35, 20)
    points_5 = (40, 17, 39)
    points_15 = (15, 41)
    points_20 = (14, 42)
    points_30 = (11, 45, 16)
    points_50 = (10, 12, 13, 43, 44, 46)
    points_100 = (8, 48, 9, 47)
    prize_d = (18, 19, 20, 21, 35, 36, 37, 38)
    throw = 0
    payment = 1
    points = 0      
    while throw in range (num_tries + 1):
        while my_money > 0:
            my_money -= payment
            score = razzle_throw()
            throw += 1
            if prize_points <= points:
                print ("Yeah! You won the prize of", prize_value, "dollars for", prize_points, "points collected!")
                points -= prize_points
                my_money += prize_value
            else:
                if score in green:
                    prize_value += 100
                elif score in points_5:
                    points += 5
                    #print("Score!")
                elif score in points_15:
                    points += 15
                    #print("Score!")
                elif score in points_20:
                    points += 20
                    #print("Score!")
                elif score in points_30:
                    points += 30
                    #print("Score!")
                elif score in points_50:
                    points += 50
                    #print("Score!")
                elif score in points_100:
                    points += 100
                    #print("Score!")
                print("throw:", throw,"Score:", score, "Money:", my_money, "Points:", points, "Payment:", payment, "Prize value:", prize_value)
                if score == 29:
                    payment = payment*2
        else:
            return print ("You're broke!")
```
